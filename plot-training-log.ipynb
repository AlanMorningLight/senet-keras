```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./logs/training.log")
```

```python
df.head(2)
```

```python
df.columns = ['epoch', 'train_acc', 'train_loss', 'lr', 'val_acc', 'val_loss']
```

```python
plt.plot('epoch', 'train_acc', data=df, color='blue', linewidth=2)
plt.plot('epoch', 'val_acc', data=df, color='red', linewidth=2)
plt.legend()
plt.title("Accuracy Plot")
plt.xlabel("epoch")
plt.ylabel("accuracy")

plt.savefig('./images/plot-accuracy.png')
```

```python
plt.plot('epoch', 'train_loss', data=df, color='blue', linewidth=2, linestyle='dashed')
plt.plot('epoch', 'val_loss', data=df, color='red', linewidth=2, linestyle='dashed')
plt.legend()
plt.title("Loss function Plot")
plt.xlabel("epoch")
plt.ylabel("loss")

plt.savefig('./images/plot-loss.png')
```
